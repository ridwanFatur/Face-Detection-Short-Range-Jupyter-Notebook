```python
import numpy as np
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
from scipy.special import expit
```

## Preprocess

```python
def preprocess_data(image_path, input_size):
    image = Image.open(image_path).convert("RGB")
    image_resized = image.resize(input_size)
    image_np = np.array(image_resized).astype(np.float32)
    image_np = image_np / 255
    input_data = np.expand_dims(image_np, axis=0)
    return input_data
```

## Inference

```python
def run_inference(interpreter):
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    interpreter.set_tensor(input_details[0]["index"], input_data)
    interpreter.invoke()
    regressors = interpreter.get_tensor(output_details[0]["index"])
    classificators = interpreter.get_tensor(output_details[1]["index"])
    return regressors, classificators
```

## Postprocess

```python
def postprocess_data(regressors, classificators, input_size, confident_threshold = 0.5):
    def generate_anchors_mediapipe():
        anchors = []
        feature_map_1_size = 16
        for i in range(feature_map_1_size):
            for j in range(feature_map_1_size):
                cx = (j + 0.5) / feature_map_1_size
                cy = (i + 0.5) / feature_map_1_size
                for anchor_id in range(2):
                    if anchor_id == 0:
                        w = 0.1484375
                        h = 0.1484375
                    else:
                        w = 0.75
                        h = 0.75
                    anchors.append([cx, cy, w, h])
        feature_map_2_size = 8
        for i in range(feature_map_2_size):
            for j in range(feature_map_2_size):
                cx = (j + 0.5) / feature_map_2_size
                cy = (i + 0.5) / feature_map_2_size
                anchor_sizes = [0.1484375, 0.2109375, 0.2734375, 0.3359375, 0.3984375, 0.75]
                for size in anchor_sizes:
                    anchors.append([cx, cy, size, size])
        return np.array(anchors)
    
    # Prepare anchors
    anchors = generate_anchors_mediapipe()
        
    def get_valid_regressors():
        conf_logits = classificators[0, :, 0]
        conf_scores = expit(conf_logits)
        valid_detections = conf_scores > confident_threshold
        
        valid_regressors = regressors[0][valid_detections]
        valid_indices = np.where(valid_detections)[0]       
        return valid_regressors, valid_indices
        
    def decode_single_face(single_regressor, anchor_idx, anchors):
        anchor = anchors[anchor_idx]
        anchor_cx, anchor_cy, anchor_w, anchor_h = anchor
        dx, dy, dw, dh = single_regressor[:4]
        decoded_cx = dx / input_size + anchor_cx
        decoded_cy = dy / input_size + anchor_cy
        decoded_w = dw / input_size
        decoded_h = dh / input_size
        x1 = (decoded_cx - decoded_w / 2) * input_size
        y1 = (decoded_cy - decoded_h / 2) * input_size
        x2 = (decoded_cx + decoded_w / 2) * input_size
        y2 = (decoded_cy + decoded_h / 2) * input_size
        x1 = np.clip(x1, 0, input_size)
        y1 = np.clip(y1, 0, input_size)
        x2 = np.clip(x2, 0, input_size)
        y2 = np.clip(y2, 0, input_size)
        keypoints = []
        for i in range(6):
            kp_x = single_regressor[4 + i*2]
            kp_y = single_regressor[4 + i*2 + 1]
            decoded_kp_x = (kp_x / input_size + anchor_cx) * input_size
            decoded_kp_y = (kp_y / input_size + anchor_cy) * input_size
            decoded_kp_x = np.clip(decoded_kp_x, 0, input_size)
            decoded_kp_y = np.clip(decoded_kp_y, 0, input_size)
            keypoints.append([decoded_kp_x, decoded_kp_y])
        return {
            'box': [x1, y1, x2, y2],
            'keypoints': keypoints,
            'box_center': [decoded_cx * input_size, decoded_cy * input_size],
            'box_size': [decoded_w * input_size, decoded_h * input_size]
        }
        
    def compute_iou(box1, box2):
        x1 = max(box1[0], box2[0])
        y1 = max(box1[1], box2[1])
        x2 = min(box1[2], box2[2])
        y2 = min(box1[3], box2[3])
        if x2 <= x1 or y2 <= y1:
            return 0.0
        intersection = (x2 - x1) * (y2 - y1)
        area1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
        area2 = (box2[2] - box2[0]) * (box2[3] - box2[1])
        union = area1 + area2 - intersection
        return intersection / union if union > 0 else 0.0
        
    def nms(results, iou_threshold=0.5):
        if len(results) == 0:
            return []
        if 'score' in results[0]:
            results = sorted(results, key=lambda x: x['score'], reverse=True)
        else:
            results = sorted(results, key=lambda x: (x['box'][2] - x['box'][0]) * (x['box'][3] - x['box'][1]), reverse=True)
        keep = []
        for i, result in enumerate(results):
            keep_current = True
            for kept_result in keep:
                iou = compute_iou(result['box'], kept_result['box'])
                if iou > iou_threshold:
                    keep_current = False
                    break
            if keep_current:
                keep.append(result)
        return keep    
    
    # Get Valid Output
    valid_regressors, valid_indices = get_valid_regressors()

    # Get Configs
    results = []
    for index in range(len(valid_indices)):
        anchor_idx = valid_indices[index]
        regressor = valid_regressors[index]
        keypoints_raw = regressor[4:].reshape(6, 2)
        result = decode_single_face(regressor, anchor_idx, anchors)
        results.append(result)
        
    # Apply NMS
    filtered_results = nms(results, iou_threshold=0.5)

    return filtered_results    
```

## Display

```python
def show_image_box(results):  
    image_edited = Image.open(image_path).convert("RGB").resize(input_size)
    for result in results:
        x1, y1, x2, y2 = result['box']
        draw = ImageDraw.Draw(image_edited)
        draw.rectangle((x1, y1, x2, y2), outline="red", width=2)
        if 'keypoints' in result:
            keypoint_colors = ["blue", "green", "yellow", "orange", "purple", "cyan"]
            for i, keypoint in enumerate(result['keypoints']):
                kp_x, kp_y = keypoint
                color = keypoint_colors[i % len(keypoint_colors)]
                draw.rectangle((kp_x-1.5, kp_y-1.5, kp_x+1.5, kp_y+1.5), outline=color, width=2)        
    return image_edited
```

## Execute

```python
model_path = "face_detection_short_range.tflite"
image_path = "sample_face.jpg"
```

```python
interpreter = tf.lite.Interpreter(model_path=model_path)
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()

input_size = (input_details[0]['shape'][1], input_details[0]['shape'][1])
```

```python
input_data = preprocess_data(image_path, input_size)
```

```python
regressors, classificators = run_inference(interpreter)
```

```python
output_data = postprocess_data(regressors, classificators, input_size[0])
```

```python
show_image_box(output_data)
```
